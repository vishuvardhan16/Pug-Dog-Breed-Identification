```python
from keras.applications.resnet50 import ResNet50,preprocess_input,decode_predictions
from keras.preprocessing import image
from skimage.io import imread

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
model = ResNet50(weights='imagenet')
```

```python
img = imread('D:\\Data Science\\Pug-Dog-Breed-Identification\\pug.jfif')
```

```python
plt.imshow(img)
plt.show()
```

```python
img = image.load_img ( 'D:\\Data Science\\Pug-Dog-Breed-Identification\\pug.jfif' , target_size = ( 224 , 224 ) )
x = image.img_to_array ( img )
x = np.expand_dims ( x, axis=0 )
preds = model.predict(x)
```

```python
h = decode_predictions(preds)[-1]
```

```python
for i in h:
    print( i[1] , str(round((i[2]*100),4))+' %' )
```
